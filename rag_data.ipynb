```python
import re

# 주어진 system_prompt에서 Context 부분을 추출하는 함수
def extract_contexts(system_prompt):
    # CONTEXT="""...""" 사이에 있는 텍스트 추출
    context_match = re.search(r'CONTEXT="""\n(.*?)"""', system_prompt, re.DOTALL)
    if not context_match:
        return []

    context_text = context_match.group(1)  # 추출한 전체 Context 텍스트

    # (context n)= ... 형식으로 개별 Context 텍스트를 추출
    contexts = re.findall(r'\(context \d+\)=(.*?)(?=\n\(context \d+\)=|\Z)', context_text, re.DOTALL)

    # 각 Context에서 "Title:"로 시작하는 줄 제거
    cleaned_contexts = []
    for context in contexts:
        lines = context.split('\n')
        cleaned_lines = [line for line in lines if not line.strip().startswith('Title:')]
        cleaned_context = '\n'.join(cleaned_lines).strip()
        cleaned_contexts.append(cleaned_context)

    return cleaned_contexts

system_prompt = df.iloc[0]['system']  # 데이터프레임에서 첫 번째 'system' 텍스트 가져오기

# Context 추출 및 출력
extracted_contexts = extract_contexts(system_prompt)
for i, context in enumerate(extracted_contexts, 1):
    print(f"Context {i}:")
    print(context)
    print("-" * 50)
```

```python
# 각 행의 'system' 열에서 Context를 추출하여 contexts 리스트에 저장
contexts = []
for i in range(len(df)):
    system_prompt = df.iloc[i]['system']  # 현재 행의 'system' 텍스트 가져오기
    extracted_contexts = extract_contexts(system_prompt)  # Context 추출
    contexts.append(extracted_contexts)  # 추출된 Context를 contexts 리스트에 추가
```

```python
# 각 행에서 필요한 정보를 리스트로 추출
questions = [df.iloc[i]['human'] for i in range(len(df))]  # 'human' 열에서 질문 리스트 생성
contexts_answers_idxs = [df.iloc[i]['answer_position'] - 1 for i in range(len(df))]  # 'answer_position'에서 인덱스 조정 후 리스트 생성
contexts_answers = [df.iloc[i]['answer_context_summary'] for i in range(len(df))]  # 'answer_context_summary' 열에서 답변 요약 리스트 생성
answers = [df.iloc[i]['answer'] for i in range(len(df))]  # 'answer' 열에서 실제 답변 리스트 생성
```

```python
import pickle

# RAG 데이터 구조 생성
rag_data = {
    'questions': questions,             # 질문 리스트
    'contexts': contexts,               # 각 질문에 대한 전체 Context 리스트
    'contexts_answer_idx': contexts_answers_idxs,  # 각 질문에 대한 정답 위치 인덱스 리스트
    'contexts_answers': contexts_answers,  # 각 질문에 대한 Context 요약 리스트
    'answers': answers                  # 각 질문에 대한 실제 답변 리스트
}

# rag_data 객체를 파일에 저장
with open('./res/rag_data.pkl', 'wb') as f:
    pickle.dump(rag_data, f)
```
